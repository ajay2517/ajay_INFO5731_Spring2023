# **Question 2**

(30 points). Write a python program to **clean the text data** you collected above and save the data in a new column in the csv file. The data cleaning steps include:

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the [stopwords list](https://gist.github.com/sebleier/554280).

(4) Lowercase all texts

(5) Stemming. 

(6) Lemmatization.

```python
# Write your code here

# importing text into a df
import pandas as pd
data_url="https://raw.githubusercontent.com/ajay2517/ajay_INFO5731_Spring2023/main/data.csv"
df = pd.read_table(data_url,names=['text'])
df


```

```python
#Number of characters
df['Number of Characters'] = df['text'].str.len()
df
```

```python
#Number of numerics
df['Numbers of numerics'] = df['text'].apply(lambda y: len([y for y in y.split() if y.isdigit()]))
df
```

```python
#digit removal
df['After digits removal'] = df['text'].apply(lambda y: ''.join([j for j in y if not j.isdigit()]))
df
```

```python
#Stopwords removal
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
st = stopwords.words('english')
df['Stopwords Removal'] = df['text'].apply(lambda y: " ".join(y for j in y.split() if y not in st))
df
```

```python
#Lower casing
df['Lower Case'] = df['text'].apply(lambda y: " ".join(y.lower() for j in y.split()))
df
```

```python
#Stemming
from nltk.stem import PorterStemmer
s = PorterStemmer()
df['After Stemming'] = df['Lower Case'].apply(lambda y: " ".join([s.stem(w) for w in y]))
df
```

```python
#Punctuation removal
df['Removal of Punctuation'] = df['Lower Case'].str.replace('[^\w\s]','')
df
```

```python
#Lemmatization
import nltk
nltk.download('omw-1.4')
from textblob import Word
import nltk
nltk.download('wordnet')
df['After Lemmatization'] = df['Removal of Punctuation'].apply(lambda y: " ".join([Word(w).lemmatize() for w in y.split()]))
df
```

```python
df.to_csv('cleaned_data.csv',index=False)
```
